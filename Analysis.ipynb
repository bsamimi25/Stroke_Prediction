```python
import psycopg2
import pandas as pd

# Import password
from config import password
```

```python
# Connect to Postgresql

con= psycopg2.connect(
            host="127.0.0.1",
            database= "stroke_db",
            user= "postgres",
            password=password)
```

```python
cur = con.cursor()

# execute query 
cur.execute("select app_id, gender, age, hypertension,heart_disease,ever_married,work_type,residence_type,avg_glucose_level,bmi,smoking_status,stroke from total_stroke")

rows=cur.fetchall()

rows
```

```python
stroke_df= pd.DataFrame(rows, columns= ["app_id", "gender", "age", "hypertension","heart_disease", 
                                        "ever_married", "work_type", "residence_type", "avg_glucose_level", 
                                        "bmi","smoking_status", "stroke"])

stroke_df.head()
```
